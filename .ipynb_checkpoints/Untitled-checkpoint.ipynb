```python
tn_movie_budgets = pd.read_csv('Data/tn.movie_budgets.csv')
tn_movie_budgets.head()
```

```python
budgets= tn_movie_budgets['production_budget']
domestic = tn_movie_budgets['domestic_gross']
worldwide = tn_movie_budgets['worldwide_gross']

def dollar_to_int(dollar):
    dollar = dollar.replace(',','')
    dollar = int(dollar.replace('$',''))
    return dollar

budgets = budgets.apply(dollar_to_int)
domestic = domestic.apply(dollar_to_int)
worldwide = worldwide.apply(dollar_to_int)
```

```python
sns.scatterplot(budgets, domestic, color = 'r')

m,b = np.polyfit(budgets, domestic, 1)
yp = np.polyval([m, b], budgets)
plt.plot(budgets,yp)
plt.show()
print('m:',m,'b:',b)
print(m * 1000000 + b)
```

```python
sns.scatterplot(budgets, worldwide, color = 'g')
n, c = np.polyfit(budgets, worldwide, 1)
yp2 = np.polyval([n, c], budgets)
plt.plot(budgets,yp2)


plt.show()
print('m:',n,'b:',c)
print(1000000 * n + c)
```

```python
rt_movies = pd.read_table('Data/rt.movie_info.tsv')
rt_movies.head()
```

```python
rt_movies.dropna(subset=['genre'],inplace=True)
genre_counts = {}
for genrelist in rt_movies['genre']:
    genrelist = genrelist.split('|')
    for genre in genrelist:
        if genre not in genre_counts:
            genre_counts[genre] = 1
        else:
            genre_counts[genre] += 1

genre_counts
genres = list(genre_counts.keys())
```

```python
genres
```

```python
sorted_genres = dict(sorted(genre_counts.items(), key=lambda x: x[1], reverse=True))
sorted_genres
```

```python
import itertools
top10 = dict(itertools.islice(sorted_genres.items(), 10)) 
top10
```

```python
top10genres = sns.barplot(list(top10.values()), list(top10.keys()))
plt.show()
```

```python
tmdb_movies['release_date'] = pd.to_datetime(tmdb_movies['release_date'])

tmdb_movies['release_date'] == max(tmdb_movies['release_date'])
```

```python
def month_count(dates,month_number):
    movies_in_month = []
    for date in dates:
        if date.month==month_number:
            movies_in_month.append(date)
    return len(movies_in_month)

months_and_movies = {}
for m in range(1,13):
    months_and_movies[m] = month_count(tmdb_movies['release_date'],m)
    
months_and_movies
```

```python
x = list(months_and_movies.keys())
y = list(months_and_movies.values())

sns.barplot(x,y)
plt.show()
```

```python
bom = pd.read_csv('Data/bom.movie_gross.csv')
bom.head()
```

```python
bom.shape
```

```python
bom = bom.dropna(subset = ['studio','domestic_gross', 'foreign_gross'])

def str_to_int(entry):
    if ',' in entry:
        entry = entry.replace(',','')
        entry = float(entry)
    else:
        entry = float(entry)
    return entry

bom['foreign_gross'] = [str_to_int(e) for e in bom.foreign_gross]
```

```python
bom['foreign_gross']
```

```python
bom.shape
```

```python
studios = list(bom['studio'].unique())
```

```python
my_list = list(bom.studio.value_counts())
```

```python
bv = bom.groupby('studio').get_group('BV')
```

```python
bvyears = list(bv.year.unique())
gross_by_year = {}
for year in bvyears:
    gross_by_year[str(year)] = 0
gross_by_year
```

```python
for year in bvyears:
    gross_by_year[str(year)] = sum(bv.groupby('year').get_group(year)['domestic_gross'])
```

```python
gross_by_year
```

```python
def get_gross_info(data,gross_string):
    gross_info = {}
    for studio in studios:
        studio_df = data.groupby('studio').get_group(studio)
        years = list(studio_df.year.unique())
        gross_by_year = {}
        for year in years:
            total = sum(studio_df.groupby('year').get_group(year)[gross_string])
            gross_by_year[str(year)] = total
        gross_info[studio] = gross_by_year
    return(gross_info)
```

```python
domestic_gross_info = get_gross_info(bom, 'domestic_gross')
```

```python
foreign_gross_info = get_gross_info(bom, 'foreign_gross')
```

```python
foreign_gross_info['BV']
```

```python
for studio in studios:
    x = list(foreign_gross_info[studio].keys())
    y = list(foreign_gross_info[studio].values())
    a = list(domestic_gross_info[studio].keys())
    b = list(domestic_gross_info[studio].values())
    fig, (ax1,ax2)= plt.subplots(2,1,constrained_layout=True, sharey=True)
    ax1.set_title('foreign')
    ax2.set_title('domestic')
    ax1.bar(x, y)
    ax2.bar(a, b, color = 'r')
    fig.suptitle(studio)
    plt.show()
```
